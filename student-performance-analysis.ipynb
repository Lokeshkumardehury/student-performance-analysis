```python
import matplotlib.pyplot as plt

df['G3'].hist()

plt.title('Final Grade Distributions')
plt.show()
```

```python
df.groupby('studytime')['G3'].mean().plot(kind='bar',ylabel='Marks')

plt.title('Study Time vs Marks')

plt.show()
```

```python
df.groupby("sex")["G3"].mean().plot(kind="bar", ylabel='marks')
plt.title("Gender vs Marks")
plt.show()
```

```python
top_students = df[df["G3"] > 15]
top_students
```

```python
plt.scatter(df["absences"], df["G3"])
plt.xlabel("Absences")
plt.ylabel("Final Grade")
plt.title("Absences vs Marks")
plt.show()
```
